### Exploratory Data Analysis 

```python
#Most selling products
prod_sales = pd.DataFrame(df.groupby('Item').sum()['Sales Quantity'])
prod_sales.sort_values(by = ['Sales Quantity'], inplace=True, ascending=False)
prod_sales[:10]
```

```python
plt.figure(figsize=(16,5))
plt.plot(df_temp['DateKey'], df_temp['Sales Amount'])
plt.xticks(rotation = 'vertical', size = 12)
plt.title('OVERALL SALES TREND', size = 15)
plt.show()
```

```python
plt.figure (figsize=(10,8))
sns.countplot(x='U/M', data=df)
plt.show()
```

```python
x = df.drop_duplicates(subset = ["Item"])
print(x.shape)
prod_sales= pd.DataFrame(x.groupby('Item').sum()['List Price'])
prod_sales.sort_values(by = ['List Price'], inplace = True, ascending = False)
y = prod_sales[:20]
y 
```
